```python
import yfinance as yf

# 创建特斯拉股票的 Ticker 对象
tesla = yf.Ticker("TSLA")

# 提取特斯拉股票历史数据
tesla_data = tesla.history(period="max")

# 重置索引并显示前 5 行
tesla_data.reset_index(inplace=True)
print(tesla_data.head())

# 保存为 CSV（可选）
tesla_data.to_csv("tesla_stock_data.csv", index=False)
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# 特斯拉收入数据页面
url = "https://www.macrotrends.net/stocks/charts/TSLA/tesla/revenue"

# 模拟浏览器访问
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/96.0.4664.110 Safari/537.36"
}
response = requests.get(url, headers=headers)
soup = BeautifulSoup(response.text, "html.parser")

# 查找收入表格
tables = soup.find_all("table")
if len(tables) > 1:
    revenue_table = tables[1]
    rows = revenue_table.find_all("tr")

    # 提取收入数据
    tesla_revenue = []
    for row in rows[1:]:
        cols = row.find_all("td")
        date = cols[0].text.strip()
        revenue = cols[1].text.strip().replace("$", "").replace(",", "")
        tesla_revenue.append([date, revenue])

    # 转换为 DataFrame
    tesla_revenue_df = pd.DataFrame(tesla_revenue, columns=["Date", "Revenue"])
    print(tesla_revenue_df.head())

    # 保存为 CSV（可选）
    tesla_revenue_df.to_csv("tesla_revenue.csv", index=False)
else:
    print("未找到收入表格！")
```

```python
# 创建 GameStop 股票的 Ticker 对象
gamestop = yf.Ticker("GME")

# 提取 GameStop 股票历史数据
gamestop_data = gamestop.history(period="max")

# 重置索引并显示前 5 行
gamestop_data.reset_index(inplace=True)
print(gamestop_data.head())

# 保存为 CSV（可选）
gamestop_data.to_csv("gamestop_stock_data.csv", index=False)
```

```python
# GameStop 收入数据页面
url = "https://www.macrotrends.net/stocks/charts/GME/gamestop/revenue"

# 模拟浏览器访问
response = requests.get(url, headers=headers)
soup = BeautifulSoup(response.text, "html.parser")

# 查找收入表格
tables = soup.find_all("table")
if len(tables) > 1:
    revenue_table = tables[1]
    rows = revenue_table.find_all("tr")

    # 提取收入数据
    gamestop_revenue = []
    for row in rows[1:]:
        cols = row.find_all("td")
        date = cols[0].text.strip()
        revenue = cols[1].text.strip().replace("$", "").replace(",", "")
        gamestop_revenue.append([date, revenue])

    # 转换为 DataFrame
    gamestop_revenue_df = pd.DataFrame(gamestop_revenue, columns=["Date", "Revenue"])
    print(gamestop_revenue_df.head())

    # 保存为 CSV（可选）
    gamestop_revenue_df.to_csv("gamestop_revenue.csv", index=False)
else:
    print("未找到收入表格！")
```

```python
import matplotlib.pyplot as plt

# 绘制特斯拉股票价格和收入对比
fig, ax1 = plt.subplots()

# 股票价格
ax1.set_xlabel("Date")
ax1.set_ylabel("Stock Price", color="tab:blue")
ax1.plot(tesla_data["Date"], tesla_data["Close"], label="Stock Price", color="tab:blue")
ax1.tick_params(axis="y", labelcolor="tab:blue")

# 收入数据
ax2 = ax1.twinx()
ax2.set_ylabel("Revenue", color="tab:red")
ax2.plot(pd.to_datetime(tesla_revenue_df["Date"]), tesla_revenue_df["Revenue"], label="Revenue", color="tab:red")
ax2.tick_params(axis="y", labelcolor="tab:red")

fig.tight_layout()
plt.title("Tesla Stock Price and Revenue")
plt.show()
```
